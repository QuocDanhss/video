# Khai báo các thư viện cần thiết

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

# Loading dữ liệu lên jupiter

```python
data = pd.read_excel('CuoiKy.xlsm')
```

# Lưu dữ liệu vào dataframe

```python
CustomerMaster = pd.read_excel('CuoiKy.xlsm', sheet_name = 'Customer_Master')
```

# Tạo ndarray sử dụng numpy, 2D

```python
column1_values = CustomerMaster['Longitude'].values
column2_values = CustomerMaster['Latitude'].values

# Create a 2D array from the two columns
data = np.column_stack((column1_values, column2_values))
data.shape
```

# Sử dụng phương pháp Elbow tìm số cụm phù hợp

```python
# Generate WCSS values for different numbers of clusters
wcss = []
for k in range(1, 12):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)

# Plot the WCSS values against the number of clusters
plt.plot(range(1, 12), wcss)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

# --> Chọn 7 cụm
Thực hiện Kmean trên dữ liệu longitude và latitude của khách hàng

```python
kmeans = KMeans(n_clusters=7)
kmeans.fit(data)

kmeans = KMeans(init="k-means++", n_clusters=7, n_init=4)
kmeans.fit(data)
```

# Vẽ biểu đồ

```python
# Step size of the mesh. Decrease to increase the quality of the VQ.
h = 0.02  # point in the mesh [x_min, x_max]x[y_min, y_max].

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in the mesh. Use last trained model.
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(
    Z,
    interpolation="nearest",
    extent=(xx.min(), xx.max(), yy.min(), yy.max()),
    cmap=plt.cm.Paired,
    aspect="auto",
    origin="lower",
)

plt.plot(data[:, 0], data[:, 1], "k.", markersize=2)

# Plot the centroids as numbers
centroids = kmeans.cluster_centers_
cluster_numbers = range(1, len(centroids) + 1)
for cluster_number, centroid in zip(cluster_numbers, centroids):
    plt.text(
        centroid[0],
        centroid[1],
        str(cluster_number),
        color="red",
        fontsize=12,
        fontweight="bold",
        horizontalalignment="center",
        verticalalignment="center",
    )

plt.title("K-means clustering on the dataset\n" "Centroids are marked with cluster numbers")
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```
